# Load SHAP values

````python
shap_values = object_load('../outputs/SHAP/cumulative.pickle')
````

# Fig

````python
WIDTH = 8
HEIGHT = 4
````

## A

````python
plt.close('all')

shap.plots.bar(shap_values, show=False)

f = plt.gcf()

ax = plt.gca()

f.set_figwidth(WIDTH)
f.set_figheight(HEIGHT)

f.patch.set_facecolor('white')


labels=[
    r'$M_s$',
    r'$\nu$',
    r'$M_d$',
    r'$Y$',
    r'$\mu$',
    r'$b$',
    r'$\omega$',
    r'$M_p$',
]*2

ax.set_yticklabels(labels)


ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)

# ax.set_xlim([0, 0.1])

f.tight_layout(rect=[0.028, 0, 0.9, 0.92])

plt.annotate(
    'A',
    (0.05, 0.9),
    xycoords='figure fraction',
    color=(0.4, 0.4, 0.4),
    size=20,
)
````

````python
f.savefig('../figures/to_combine/app5_fig5_1.jpg', dpi=1200)
````

## B

````python
plt.close('all')

shap.plots.beeswarm(shap_values, plot_size=(WIDTH,HEIGHT), show=False)

f = plt.gcf()

ax = plt.gca()

f.patch.set_facecolor('white')


ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

labels=[
    r'$M_p$',
    r'$\omega$',
    r'$b$',
    r'$\mu$',
    r'$Y$',
    r'$M_d$',
    r'$\nu$',
    r'$M_s$',
]

ax.set_yticklabels(labels)

f.tight_layout(rect=[0, 0, 1, 0.95])

plt.annotate(
    'B',
    (0.05, 0.9),
    xycoords='figure fraction',
    color=(0.4, 0.4, 0.4),
    size=20,
)
````
